Building Functional Mode

```python
from tensorflow.keras.models import Model
```

```python
from tensorflow.keras.layers import Input,Flatten,Dense
```

```python
ip=Input(shape =(32,32,3))
```

```python
flat=Flatten()(ip)
```

```python
Den1=Dense(64,activation="relu")(flat)
```

```python
Den2=Dense(128,activation="relu")(Den1)
```

```python
Den3=Dense(256,activation="relu")(Den2)
```

```python
out=Dense(10,activation="softmax")(Den3)
```

```python
fun_model= Model(inputs=ip,outputs=out)
```

```python
fun_model
```

```python
fun_model.summary()
```

Compile

```python
fun_model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
fun_model
```

```python
fun_model.fit(X_train_scale,y_train,epochs=10,batch_size=64)
```

```python
fun_model.evaluate(X_test_scale,y_test)
```
